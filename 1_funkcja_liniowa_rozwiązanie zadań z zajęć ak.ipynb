```python
def linear_function (a, b, x):
    return a*x + b
```

```python
a=2
b=4
x=7
y= linear_function (a, b, x)
```

```python
print(f'y= {a}*{x} + {b} = {y}')
```

```python
narysować wykres kilku fukncji liniowych  o różnych współczynnikach a (slope) zamknij w funkjci
```

```python
a=[-2, -1, 0, 1, 2] # definiuje różne współczynniki a slope
b= 1
```

```python
def draw_line_func_slope(a,b, x):
    y=linear_function(a,b,x)
    plt.plot(x,y, label=str(a))
```

```python
for ai in a:
    draw_line_func_slope(ai, b, x)
plt.legend()
plt.title('Funkcja liniowa z różnymi wartościami współczynnika slope')
plt.show()
```

```python
narysować wykres kilku fukncji liniowych  o różnych współczynnikach b (intercept) zamknij w funkjci
```

```python
a=1
b=[-2, -1, 0, 1, 2] # definiuje różne współczynniki intercept tj różne wartości współczynnika
```

```python
def draw_line_funkction_intercept(a,b,x):
    y=linear_function(a, b,x)
    plt.plot(x,y, label=str(b))
```

```python
for bi in b:
    draw_line_funkction_intercept(a,bi,x)
plt.legend()
plt.title('Funkcja liniowa z różnymi wartościami współczynnika intersept')
plt.show()
```

```python
odległość miedzy dwiema funcjaki o różnych wartościach współczynnika slope
```

```python
x= np.arange(-10,5,1) #Zwraca równomiernie rozmieszczone wartości w przedziale {-10,10)
x # arrayka od x
```

```python
y1=linear_function(1,2,x)
y2=linear_function(2,2,x)
yerr=y2-y1
```

```python
print(y1)
print(y2)
print(yerr)
```

```python
fig=plt.figure(figsize=(10,10))
plt.plot(x,y1)
plt.plot(x, y2)
lolims = np.greater(yerr, 0)
uplims = np.less(yerr, 0)
plt.errorbar(x,y1,yerr=np.abs(yerr), label='both limit (default)', lolims=lolims, uplims=uplims)
```
