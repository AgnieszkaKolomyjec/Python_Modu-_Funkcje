```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
# losowanie zestawu danych. Założenie. Zestaw danych ilustruje 100 klientów i ich zwyczaje zakupowe
x= np.random.normal(3, 1, 100)      #x  liczba minut przed dokonaniem zakupu.
y = np.random.normal(150, 40, 100) / x   #y kwota pieniędzy wydaną na zakup.
```

```python
plt.scatter(x, y, marker='o')
plt.title('Klientci w sklepie stacjonarnym i ich zwyczaje zakupowe')
plt.xlabel('Liczba minut przed dokonaniem zakupu')
plt.ylabel('Kwotę pieniędzy wydaną na zakup')

plt.show()
```

```python
#podział zbioru na część testową i treningową
X_train = x[:80]    #losowy wybór 80% oryginalnych danych.
y_train = y[:80] 
X_test = x[80:]     # zestaw testowy stanowi pozostałe 20%
y_test = y[80:]
```

```python
# Błąd 'obiektu „numpy.ndarray” nie jest wywoływalny' dlatego muszę dodać poly1d -> Jednowymiarowa klasa wielomianowa.

model = np.poly1d(np.polyfit(X_train, y_train, deg=3)) #Dopasowanie wielomianowe metodą najmniejszych kwadratów. Deg =Stopień dopasowania wielomianu
```

```python
model
```

```python
line = np.linspace(0, 5, 100)
```

```python
plt.scatter(X_train, y_train)
plt.plot(line, model(line))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_train, model(X_train))

print(r2)
```

```python
r2_test = r2_score(y_test, model(X_test))
print(r2_test)
```

```python
Model pasuje do danych treningowych, na danych testowych wypada gorzej. Raczej nie użyłabym go do przewidywania wartości
```
