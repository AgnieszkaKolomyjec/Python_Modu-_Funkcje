```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
plt.scatter(x,y)
```

```python
mymodel=np.polyfit(x,y,deg=2)
```

```python
mymodel
```

```python
mymodel=np.poly1d(np.polyfit(x,y,deg=2))
```

```python
mymodel
```

```python
myline=np.linspace(1,25,100) #sposób wyświetlania linii, zaczynamy od pozycji 1 i kończymy na pozycji 25
```

```python
plt.scatter(x,y)
plt.plot(myline, mymodel(myline))
```
